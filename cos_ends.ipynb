```python
import os
group1="3 extended cos ends"
group2="Lambda-like (5 cos ends)"
group3="P2-like (5 extended cos ends)"
group4="N15-like (cos ends)"

three=[]
five=[]
##use this script to pull out different categories of phages i wish to work with

ktblast=open("ktblast_terminase_IDs.txt").readlines()
whatsleft=open("whats_left.txt").read().split("\n")

categories={}

grouplist=[]

for line in ktblast[1:]:
    name=line.split(",")[0].split("_")[0]
    cat=line.split(",")[2].replace("\n","")
    if name in whatsleft:
        categories[name]=cat

for value in categories.keys():
    if categories.get(value)==group1 or categories.get(value)==group2 or categories.get(value)==group3 or categories.get(value)==group4:
        grouplist.append(value)
print(grouplist)

for value in categories.keys():
    if categories.get(value)==group1 or categories.get(value)==group4:
        three.append(value)
    elif categories.get(value)==group2 or categories.get(value)==group3:
        five.append(value)
```

```python
%%R -i grouplist

cos5_pattern=vector()
cos5_splitpoint=vector()
other_pattern=vector()
for (item in grouplist){
    #print(item)
    filename<-paste(item,"step1.contig1.genomecoverage.txt",sep="")
    coverage<-read.table(filename, sep="\t")
    mean_cov<-mean(coverage$V3)
    min_cov<-min(coverage$V3)
    minlocation<-round(median(coverage$V2[coverage$V3==min(coverage$V3)]))
    #print(minlocation)
    if (minlocation>(nrow(coverage)-1000)|minlocation<1000){
#        minlocation<-round(median(coverage$V2[coverage$V3[100:nrow(coverage)-100]==min(coverage$V3[100:nrow(coverage)-100])]))
        minlocation<-round(median(coverage$V2[coverage$V3[1000:(nrow(coverage)-1000)]==min(coverage$V3[1000:(nrow(coverage)-1000)])]))+1000
        #print(paste("for", item,"the absolute minimum is", min_cov, "and the local minimum is", min(coverage$V3[1000:nrow(coverage)-1000]),"at",minlocation,sep=" "))
        #print(coverage$V2[coverage$V3==min(coverage$V3[1000:nrow(coverage)-1000])])
        min_cov=min(coverage$V3[1000:(nrow(coverage)-1000)])
        #print("change")
        mark2="$"
    }else{
        mark2=""
    }
    if (min_cov<(mean(coverage$V3)-(sd(coverage$V3)*4))){
        mark="*"
        cos5_pattern<-c(cos5_pattern,item)
        cos5_splitpoint<-c(cos5_splitpoint, minlocation)
        #print("significant dip")
        
    } else{
        mark=""
        other_pattern=c(other_pattern, item)
    }
    
        with(coverage, plot(V2, V3, type="n",xlab="position",ylab="coverage", main=paste(item, mark2,mark, sep="")))
        with(coverage, lines(V2,V3))
        points(minlocation, min_cov, pch=19, col="red")
    
    
    }
print(cos5_pattern)
```
