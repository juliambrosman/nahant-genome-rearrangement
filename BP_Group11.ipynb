```python
%%R

sip_treated<-division$phage[division$div==1|division$div==2]
sip_sd_factor=1
sip_split<-vector()

exp<-vector()

for (item in sip_treated){
    filename<-paste(item,"step1.contig1.genomecoverage.txt",sep="")
    
    coverage<-read.table(filename, sep="\t")
       
    length<-nrow(coverage)
    omit_ends_cov<-coverage[(2000:(length-2000)),]   #excluding larger ends of the contig in this case   
    
    sip_split<-c(sip_split,omit_ends_cov$V2[omit_ends_cov$V3==max(omit_ends_cov$V3)]%>%median%>%round)
    with(coverage, plot(V2, V3, type="l", lty=2, col="purple4"))
    points(omit_ends_cov$V2[omit_ends_cov$V3==max(omit_ends_cov$V3)]%>%median%>%round, max(omit_ends_cov$V3), col="magenta", pch=10, cex=3)
    }
print(sip_split)
```

>Based on this assessment, the sip protocol mis-calls the location I'd like to cut for two phages in group1.  I'll have to re-assemble these two phages using an adjusted genome.

```python
%Rpull sip_split sip_treated

sipcut=sip_split.tolist()

for i in range(0,len(sip_split)):
    phage=sip_treated[i]
    print(phage)
    new_file=open(phage+"step1.contig1.rearrange.fasta","w")
    cut=sipcut[i]
    fastafile=phage+"step1.contig1.fasta"
    f=open(fastafile)                                       #open fasta file
    sequence=""
    for row in f:                           #define title and sequences
        if row.startswith(">"):
            title=row
        else:
            sequence+=row
    sequence=sequence.replace("\n","")
    upstream=sequence[0:int(cut)-1]                    #id sequence before start of small subunit terminase
    downstream=sequence[int(cut)-1:] 
    newseq=downstream+upstream
    
    new_file.write(">"+phage+"step1.contig1.cos.rearrange\n"+newseq)
    new_file.close()
```

```python
from pyfaidx import Fasta
from Bio.Seq import Seq

for line in grouplist:
    name=line.rstrip("\n")
    
    blastoutputname=name+"vs.terminases.txt"
    blastout=open(blastoutputname)
    blastout=blastout.read()
    vec=blastout.split("\t")
    if len(vec)>2:
        start=int(vec[6])
        inputfasta=name+"recruit2.fasta"
        f=Fasta(inputfasta)
        viromelen=len(f[0])
        threshhold=len(f[0])/2
        sequence=str(f[0])
        seq=Seq(str(sequence))
        outfasta=open(name+"oc.fasta","w")
        outfasta.write(">"+name+"\n")

        if start>threshhold:
            print(name+": must reverse sequence ")
            out_seq=seq.reverse_complement()
    
        else:
            print(name+": sequence is in correct orientation")
            out_seq=seq
        outfasta.write(str(out_seq))
        outfasta.close()
    else:
        print(name+": no blast match")
```
