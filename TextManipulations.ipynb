```python
import pyfaidx
from pyfaidx import *

fastafile=Fasta("2.130.O.reads.fasta")
titles=open("2.130.O.high_cov_read_titles.txt")
titles=titles.readlines()
```

```python
high_cov_titles=[]
for line in titles:
    title=line.replace("\n","")
    title1=title+"/1"
    title2=title+"/2"
    if title1 not in high_cov_titles:
        high_cov_titles.append(title1)
        high_cov_titles.append(title2)
```

```python
output=open("2.130.O.high_cov_reads.fasta","w")
output2=open("2.130.O.all_other_reads.fasta","w")
for line in fastafile.keys():
    sequence=str(fastafile[line])
    if line in high_cov_titles:
        output.write(">"+line+"\n"+sequence+"\n")
    else:
        output2.write(">"+line+"\n"+sequence+"\n")
output.close()
output2.close()
```
