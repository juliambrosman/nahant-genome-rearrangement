```python
group="group17"
import os

prelim_groups=open("phage_by_group.txt")
phage_grp={}
for line in prelim_groups.readlines():
    line=line.rstrip("\n")
    vec=line.split("\t")
    phage_grp[vec[0]]=vec[1]
groups=list(phage_grp.values())
unique=[]
for value in groups:
    if value not in unique:
        unique.append(value)
group17=[]
for value in phage_grp.keys():
    if phage_grp.get(value)=='CNSv2.0:17':
        group17.append(value)
group17
import os
os.chdir("./"+group)
```

```python
%%R -i group17

grouplist<-group17
#create coverage plots for each genome
for (item in grouplist){
    filename<-paste(item,"step1.contig1.genomecoverage.txt",sep="")
    coverage<-read.table(filename, sep="\t")
    with(coverage, plot(V2, V3, type="n",xlab="genome",ylab="coverage", main=item))
    with(coverage, lines(V2,V3, lty=2))
}
```

```python
%%R -i group17

mu_pattern<-vector()
mu_start<-vector()
mu_stop<-vector()
next_step<-vector()

region<-""
for (item in group17){
    
    filename<-paste(item, "step1.contig1.genomecoverage.txt", sep="")
    coverage<-read.table(filename, sep="\t")
    difference<-diff(coverage$V3)
    difference<-c(0,difference)

    cov1<-cbind(coverage, difference)
    length<-nrow(cov1)
    mid<-length/2
    
    #plot(cov1$V2, cov1$difference, main=item, col="magenta")


    cut_spot<-with(cov1[1:mid,], V2[which.max(abs(difference))])
    cut_spot2<-with(cov1[mid:length,], V2[which.min(difference)])
    if ((cut_spot2>(length-1000)|cut_spot<1000) & (length-cut_spot2>0|cut_spot>10)){
        print(paste("Mu pattern for",item, sep=" "))
        mu_pattern<-c(mu_pattern, item)
        mu_start<-c(mu_start, cut_spot)
        mu_stop<-c(mu_stop, cut_spot2)
    } else {
        print(paste("no mu pattern for",item,sep=" "))
        next_step<-c(next_step, item)
    }


}
```

```python
%%R -i group17

region<-""
for (item in group17){
    if (item!="1.028.O."){
    filename<-paste("./TentativelyFinished/",item,"s1c1.oc.genomecoverage.txt",sep="")
    }   else {
        filename<-paste(item, "directed.genomecoverage.txt", sep="")
        }
    coverage<-read.table(filename, sep="\t")
    difference<-diff(coverage$V3)
    difference<-c(0,difference)

    cov1<-cbind(coverage, difference)
    length<-nrow(cov1)
    
    maxdif<-max(abs(cov1$difference[1:length/2]))
    #mindif<-min(cov1$difference[length/2:length])
    
    cut_spot<-cov1$V2[cov1$difference==maxdif]
    #cut_spot2<-cov1$V2[cov1$difference==mindif]
    region<-paste(region, paste(item,cut_spot,sep="\t"),sep="\n")


}
region
```

```python
%Rpull region

for line in region[0].split("\n")[1:]:
    vec=line.split("\t")
    if vec[0]!="1.028.O.":
        filename="./TentativelyFinished/"+vec[0]+"s1c1.oc.trimmed.fasta"
        toopen="./TentativelyFinished/"+vec[0]+"s1c1.oc.fasta"
    else:
        filename="./TentativelyFinished/"+vec[0]+"directed.trimmed.fasta"
        toopen=vec[0]+"directed.fasta"
    print(vec[0])
    f=open(toopen)
    sequence=""
    for row in f:
        if row.startswith(">"):
            title=row
        else:
            sequence+=row
    sequence=sequence.replace("\n","")
    if vec[1]!="NA":
        output=open(filename, "w")
        start=int(vec[1])
        if vec[2]!=" ":
            stop=int(vec[2])
        else:
            stop=len(sequence)
        
        
        target=(sequence[start:stop])
        output.write(title+target)
        print(len(target))
        output.close()

    else:
        print("no peak")
```
